```python
plt.title("Spending_USD vs country")
plt.ylabel('Spending in US dollar')
plt.xlabel("Country")
sns.set_style('darkgrid')
sns.lineplot(x='Country',y='Spending_USD',data=health)

sns.despine() # to remove the spine that is outer box of the graph
plt.figure(figsize=(10,10))

```

```python
plt.title("Spending_USD vs country")
plt.ylabel('Spending in US dollar')
plt.xlabel("Country")
sns.set_style('darkgrid')
sns.barplot(x='Country',y='Spending_USD',data=health)

sns.despine() # to remove the spine that is outer box of the graph
plt.figure(figsize=(10,10))

```

```python
plt.title("Spending_USD vs country")
plt.ylabel('Spending in US dollar')
plt.xlabel("Country")
sns.set_style('darkgrid')
sns.scatterplot(x='Year',y='Spending_USD',hue='Country',data=health)

sns.despine() # to remove the spine that is outer box of the graph
plt.figure(figsize=(10,10))

```

```python
plt.title("Spending_USD vs country")
plt.ylabel('Spending in US dollar')
plt.xlabel("Country")
sns.set_style('darkgrid')
sns.scatterplot(x='Year',y='Life_Expectancy',hue='Country',data=health)

sns.despine() # to remove the spine that is outer box of the graph
plt.figure(figsize=(10,10))
```

```python
plt.title("Spending_USD vs country")
plt.ylabel('Spending in US dollar')
plt.xlabel("Country")
sns.set_style('darkgrid')
sns.scatterplot(x='Year',y='Life_Expectancy',hue='Country',data=health)


sns.despine() # to remove the spine that is outer box of the graph
plt.figure(figsize=(10,10))
```

```python
plt.rc("figure", figsize=(45 , 10))
leg=sns.FacetGrid(health,col='Country',hue='Country', height=5, aspect=.65)
leg.map(sns.lineplot,'Spending_USD','Life_Expectancy').add_legend()
```
