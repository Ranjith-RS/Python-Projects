```python
#If innings is null we cant compute the row like runs , avg, not out etc so we are going to drop it
batsman.dropna(subset=['Innings'],inplace=True)
```

```python
batsman.isnull().sum()
```

```python
batsman[batsman.isnull().any(axis=1)]
#here we can clearly saw that avearge is missed due to some mistake 
```

```python
#we are going to calculate the average in cricket
batsman['avg']=batsman['Innings']-batsman['Not Out']
batsman['Avg']=round(batsman['Runs']/batsman['avg'],2)
batsman
```

```python
#now we dont want the column named avg , so drop it
batsman.drop(batsman.loc[:, ['Average','avg']].columns, axis=1,inplace=True)
batsman
```

```python
batsman.isnull().sum()
```

```python
# still we are left with some more null values in Our DataFrome , so we are going to drop that to
batsman.dropna(subset=['Avg'],inplace=True)
batsman
```

```python
batsman.shape
```

```python
batsman.loc[batsman.Runs==0.0]
#here we have players with not even score single runs , so we are going to drop it
```

```python
#droping players with zero runs
batsman.drop(batsman[batsman['Runs'] <= 50.0].index,inplace=True)
```

```python
#change datatypes of following from float to integer
batsman[['Innings','Not Out','Runs','Hundred','Fifty','Duck Out']]=batsman[['Innings','Not Out','Runs','Hundred','Fifty','Duck Out']].astype(int)
```
